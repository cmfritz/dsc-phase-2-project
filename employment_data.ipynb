```python
# Import packages
import pandas as pd # Reading csv file 
import matplotlib.pyplot as plt # for plotting
import numpy as np
```

```python
# data sourced from US Bureau of Labor & Statistics www.bls.gov
df = pd.read_csv('data/Largest_occupations_Seattle-Bellevue-Everett_2016-2019.csv')
df['Year'] = df['Year'].astype('int64')
df = df.sort_values(by=['Year'])
df.head(3)
```

```python
# graph occupations by year
xlabels = []
fig, ax = plt.subplots(figsize=(10,8))
for key, grp in df.groupby(['Group']):
    ax.plot(grp['Year'], grp['Employment'], label=key, marker='o')
plt.title('Largest occupations in Seattle 2016 - 2019')
plt.xticks(np.arange(2016, 2020))
ax.legend(loc="best")
ax.set_xlabel('year')
ax.set_ylabel('no. of jobs')
plt.show()
```
